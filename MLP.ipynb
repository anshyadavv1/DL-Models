```python
import kagglehub
from sklearn.preprocessing import LabelEncoder
path = kagglehub.dataset_download("radheshyamkollipara/bank-customer-churn")
import os
print(path)
print(os.listdir(path))
import numpy as np
```

```python
import pandas as pd
import os

df = pd.read_csv(os.path.join(path, "Customer-Churn-Records.csv"))

print(df.head())
print(df.shape)
```

```python
df['Exited'].value_counts()
```

```python
df. drop (columns=[ 'RowNumber', 'CustomerId', 'Surname' ],inplace= True)
```

```python
df.head()
```

```python
df= pd.get_dummies(df,columns=['Geography','Gender','Card Type'],drop_first=True)
```

```python
X = df.drop (columns=['Exited' ])
y = df['Exited' ]
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model= Sequential()
model.add(Dense(5,activation='sigmoid',input_dim=17))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(X_train_scaled,y_train,epochs=20,verbose=1)
```

```python
y_log=model.predict(X_test_scaled)
```

```python
y_pred= np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
